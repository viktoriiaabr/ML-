```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import model_selection
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import graphviz

import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df['flower_species'] = df.target.apply(lambda x : iris.target_names[x])
print('Unique target values=',df['target'].unique())

df.sample(5)
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = df[['target']]
```

```python
X_train,X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size= 0.2, random_state= 1)
```

```python
model = tree.DecisionTreeClassifier(random_state= 1)
model.fit(X_train ,y_train)
```

```python
model.score(X_test, y_test)
```

```python
tree.plot_tree(model)
```

# Задача
Подобрать значения гиперпараметров модели criterion, max_depth, min_samples_leaf с помощью перебора по сетке.

Для выполнения задания необходимо сделать следующие шаги:<br>
    1. задать параметры перебора <br>
    2. осуществить перебор по сетке с кросс-валибацией <br>
    3. обучить модель <br>
    4. проверить точность на кросс-валидации <br>
    5. проверить точность на тестовой выборке <br>
    6. добавить визуализацию

```python
# 1
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 3, 4, 5, 6, None],
    'min_samples_leaf': [1, 2, 3, 4, 5]
}

# 2
grid_search = GridSearchCV(model, param_grid, cv=5)
# 3
grid_search.fit(X_train, y_train)

print("Значения гиперпараметров модели:", grid_search.best_params_)
print("Лучшая точность на кросс-валидации:", grid_search.best_score_) #4
```

```python
# 5
best_model = grid_search.best_estimator_
print("Точность на тестовой выборке:", best_model.score(X_test, y_test))
```

```python
# 6
plt.figure(figsize=(12, 8))
tree.plot_tree(best_model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
